```python
import numpy as np #numpy for random ints
import matplotlib.pyplot as plt

a = 0
b = 0
c = 0 
d = 0


arr = [a,b,c,d]
arrX = []
arrY = []


xCount = 0 
#loop for making x values from -100 to 100
while(xCount<201):
    xVals = -100
    arrX.append(xVals+xCount)
    xCount = xCount + 1

count = 0
while(count<101):
   
    arr[0] = np.random.randint(-100,100)
    arr[1] = np.random.randint(-100,100)
    arr[2] = np.random.randint(-100,100)
    arr[3] = np.random.randint(-100,100)
    youtput = (arr[2]*(arr[0]*arrX[count])+arr[3] )
    arrY.append(youtput)
    count = count + 1


plt.plot(arrY)
#plt.show()
```
